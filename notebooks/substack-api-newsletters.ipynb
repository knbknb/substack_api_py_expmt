# Usage of substack_api module


```python
from substack_api import newsletter
# from substack_api import user
from pprint import pprint
from IPython.display import display, HTML
```

```python
categories_2024 = [
    ("Knowledge, Art, Literature", ["Art & Illustration", "Comics", "Culture", "Design", "Education", "Faith & Spirituality", "Fashion & Beauty", "Fiction", "History", "Literature", "Music", "Philosophy"]),
    ("News+Politics", ["Business", "Finance", "Health Politics", "International", "News", "U.S. Politics", "World Politics"]),
    ("Science+Technology", ["Climate & Environment", "Crypto", "Health & Wellness", "Science", "Technology"]),
    ("Leisure", ["Food & Drink", "Humor", "Parenting", "Podcasts", "Sports", "Travel"])
]
```

##### List all categories on Substack:

```python
# returns a list of tupley (name, count)
categories = newsletter.list_all_categories()
categories.sort(key=lambda x: x[0])
# list categories, grouped by the 2024 categories
display(HTML('<h3>Categories (2024) and their category id</h3>'))
for cat_2024 in categories_2024:
    display(HTML(f'<h4>{cat_2024[0]}</h4>'))
    for cat in categories:
        if cat[0] in cat_2024[1]:
            #display(HTML(f'<b>{cat[0]}</b>: {cat[1]}'))
            print(f"{cat[0]}: {cat[1]}")

```

### List all `canonical_url` keys of the Newsletters

```python
n_posts = 60
newsletters_metadata_full = newsletter.get_newsletter_post_metadata(newsletter_id, start_offset=0, end_offset=n_posts)

newsletters_metadata_full.sort(key=lambda x: x["canonical_url"])
for newsletter_metadata in newsletters_metadata_full:
    print(newsletter_metadata["canonical_url"])
```

Download all the newsletters to the `./newsletters` directory.

This might take a while, and might needs to run twice, due to the rate limiting of the site.

```python
#make newsletters directory
import os
os.makedirs("newsletters", exist_ok=True)
os.chdir("/home/knut/code/git/python-stuff/substack_api_py_expmt/notebooks")
os.chdir("newsletters")
for newsletter_metadata in newsletters_metadata_full:
    print(newsletter_metadata["title"])
    outfile = f"{newsletter_metadata['slug']}.html"
    if os.path.exists(outfile):
        print(f"Skipping {outfile} - already exists")
        continue
    with open(outfile, "w") as f:
        try:
            f.write(newsletter.get_post_contents(newsletter_id, newsletter_metadata["slug"], html_only=True))
        except Exception as e:
            print(f"Error writing {outfile}: {e}")
            os.chdir("..")
os.chdir("..")
```
